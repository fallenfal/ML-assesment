### Transforming the data into numbers
Based on the data paper we will transform the values from text into integers , ex:

Child One: 2 3 1 1 3 1 2 2

were the codes drawn from table  signify:

(leaf K1) Parents occupation: pretentious (2)

(leaf K2) Childs nursery: improper (3)

(leaf K3) Form of the family: complete family (1)

(leaf K4) Number of children: (1)

(leaf K5) Housing conditions: critical (3)

(leaf K6) Financial standing of the family: convenient (1)

(leaf K7) Social conditions: slightly problematic (2)

(leaf K8) Health conditions: acceptance is recommended (2)


```python
parents = ['usual', 'pretentious' ,'great_pret']
has_nurs = ['proper' ,'less_proper', 'improper' ,'critical' ,'very_crit']
form = ['complete' ,'completed', 'incomplete', 'foster']
children = ['1', '2', '3' ,'more']
housing = ['convenient','less_conv', 'critical']
finance = ['convenient','inconv']
social = ['nonprob' ,'slightly_prob', 'problematic']
health = ['priority' ,'not_recom', 'recommended']
# we have to change class to target because the word class is reserved in python
target =  ['priority', 'not_recom' ,'recommend' ,'very_recom' ,'spec_prior']

# making a copy
df_copy = df.copy()

# Custom function to transform the data
def to_int_value(text_array, numeric_array ):
    for index, value in enumerate(text_array):
        if value in numeric_array:
            # we have to increment with one because in the paper theres no value 0, it starts from 1
            numeric_value = numeric_array.index(value) + 1
            text_array[index] = numeric_value

to_int_value(df["parents"], parents)
to_int_value(df["has_nurs"], has_nurs)
to_int_value(df["form"], form)
to_int_value(df["children"], children)
to_int_value(df["housing"], housing)
to_int_value(df["finance"], finance)
to_int_value(df["social"], social)
to_int_value(df["health"], health)
to_int_value(df["class"], target)
```

### Exploring the relation between the data and the target value

```python
sns.set_style('darkgrid')
sns.set_palette('Set2')

data_viz_copy = df.copy()

def change_parents_data(parent):
    if parent == 1:
        return 'usual'
    elif parent == 2:
        return 'pretentious'
    elif parent == 3:
        return 'great_pret'

data_viz_copy["parents"] = data_viz_copy["parents"].apply(change_parents_data)

def change_target_data(target):
    if target == 1:
        return 'priority'
    elif target == 2:
        return 'not_recom'
    elif target == 3:
        return 'recommend'
    elif target == 4:
        return 'very_recom'
    elif target == 5:
        return 'spec_prior'


data_viz_copy["class"] = data_viz_copy["class"].apply(change_target_data)

sns.set_theme(rc={'figure.figsize':(9,7)})

sns.countplot(data= data_viz_copy, x='parents',hue='class')
plt.title('Parents v/s target\n')
plt.show()

```

### Creating the Logistic Regression Model

Prepare Features and Target Variables

```python
X = df.drop("class", axis=1)
y = df["class"]
print(type(X))
print(type(y))
y = [int(label) for label in y]
```

### Split the Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print(X_train)
```

```python
from sklearn import preprocessing
normalized_X_train = preprocessing.normalize(X_train)
normalized_X_test = preprocessing.normalize(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
# notice that you can change the parameters of the model
lg_model = LogisticRegression(solver='lbfgs',max_iter=3000)
# lg_model.fit(X_train,y_train)
```

```python
lg_model.fit(X_train,y_train)
```

```python
# predicted values
y_pred = lg_model.predict(X_test)
# actual values
actual = y_test
```

```python
# create a dictionary and a data frame and check correct predictions
dic = {'Actual':actual,
       'Prediction':y_pred,
       'correct_prediction':0
       }

result  = pd.DataFrame(dic)
```

```python
# update the correct_prediction values to 1, if correct prediction
def myfunc(x,y):
    if x == y:
        return 1
    else:
        return 0
result['correct_prediction'] = result.apply(lambda x: myfunc(x.Actual, x.Prediction), axis=1)
```

```python
result[result.correct_prediction==0]
```

```python
print(f'Model Accuracy on Test Data is {len(result[result.correct_prediction==1])/len(result)}')
```

```python
from sklearn.metrics import accuracy_score
print(f'The Model Accuracy on Testing Set is {round(accuracy_score(y_test,y_pred),2)}')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
